```python
# Q1 block world problem using DFS
import copy;


initial = [['A'],['B','C'],[]];
goal = [[],[],['A','B','C']];

def genvalidstates(initial):
    validstates = [];
    for i in range(0,3):
        for j in range(0,3):
            if i!=j:
                if len(initial[i])>0:
                    temp = copy.deepcopy(initial);
                    temp[j].append(temp[i].pop());
                    validstates.append(temp);
    return validstates;



def dfs(initial,goal):
    S = [];
    discovered = [];
    S.append(initial);

    while S:
        state = S.pop();
        
        print(state);
        if state == goal:
            return state;

        if state in discovered:
            continue;

        children = genvalidstates(state);
        for child in children:
            S.append(child);

        discovered.append(state);

    return None;

print(dfs(initial,goal));
```

```python
# Q2 block world problem using BFS

initial = [['A'],['B','C'],[]];
goal = [[],[],['A','B','C']];

def genvalidstates(initial):
    validstates = [];
    for i in range(0,3):
        for j in range(0,3):
            if i!=j:
                if len(initial[i])>0:
                    temp = copy.deepcopy(initial);
                    temp[j].append(temp[i].pop());
                    validstates.append(temp);
    return validstates;
```

```python
def dfs(state, depth, max_depth, visited_states):
    print(state)

    if state == goal:
        print("Goal state reached!")
        return True

    if depth >= max_depth:
        return False

    visited_states.append(state)

    for i in range(len(state)):
        if not state[i]:
            continue

        for j in range(len(state)):
            if i == j:
                continue

            new_state = [list(tower) for tower in state]
            disk = new_state[i].pop()
            new_state[j].append(disk)

            if new_state not in visited_states:
                if dfs(new_state, depth + 1, max_depth, visited_states):
                    print("Goal state reached!")
                    return True

    return False

initial = [['B'],['C','A'],[]]
goal = [[],[],['C','B','A']]

depth_limit = 1

visited_states = []
print(f"Searching with depth limit {depth_limit}...")
if dfs(initial, 0, depth_limit, visited_states):
    print(f"Solution found with depth limit {depth_limit}!")
else:
    print(f"No solution found with depth limit {depth_limit}.")
```

```python
#UCS Problem

import heapq

Paths = {
    'S': {'A': 1, 'B': 5, 'C': 15},
    'A': {'S': 1, 'G': 10},
    'B': {'S': 5, 'G': 5},
    'C': {'S': 15, 'G': 5},
    'G': {'A': 10, 'B': 5, 'C': 5}
}

def uniform_cost_search(graph, start, goal):
    # Keep track of visited nodes
    visited = set()
    # Initialize the frontier with the start node and cost 0
    frontier = [(0, start, [])]
    
    while frontier:
        # Get the node with the lowest cost from the frontier
        cost, current, path = heapq.heappop(frontier)
        # If the goal node is reached, return the path to it
        if current == goal:
            return path + [current]
        # If the current node has not been visited yet, add it to the visited set
        if current not in visited:
            visited.add(current)
            # Add the neighbors of the current node to the frontier with their respective costs
            for neighbor, neighbor_cost in graph[current].items():
                if neighbor not in visited:
                    heapq.heappush(frontier, (cost + neighbor_cost, neighbor, path + [current]))

    # If the goal node is not reached, return None
    return None

start = 'S'
goal = 'G'
path = uniform_cost_search(Paths, start, goal)
print(path)
```
